```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
import networkx as nx
import os

# Import given number of files
# NUMBER_OF_GRAPHS_WANTED=-1, to import all files

def import_some(NUMBER_OF_GRAPHS_WANTED,path):
    # Change the directory
    os.chdir(path)
  
    # Read text File

    l = []
  
    def read_file(file_path):
        l.append(nx.read_graphml(file_path))
  
  
    # iterate through all file
    i = -1
    for file in os.listdir():
        i+=1
        if i == NUMBER_OF_GRAPHS_WANTED:
            break
        # Check whether file is in text format or not
        if file.endswith(".graphml"):
            file_path = f"{path}\{file}"
 
            # call read text file function
            read_file(file_path)
    return l
```

```python
# F(x)=P(..<x)
def diagram(path):
    graph_list = import_some(-1, path)

    l=[]
    
    graph_list.sort(key=lambda x: len(x.edges),reverse=True)
    for g in graph_list:
        l.append(1-len(g.edges)/len(graph_list[0].edges))

    for i in range(11):
        print(10*i,l[10*i])

    y = np.asarray(l)
    x = np.asarray(range(101))

    plt.plot(x,y)
    plt.show()
```
